```python
import pandas as pd
# from sklearn.feature_extraction.text import CountVectorizer
import urllib
from textblob import TextBlob, Word, Blobber 
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
clinton_debate_1 = pd.read_csv('ClintonPresDebate1.txt', sep = '\t', encoding = 'ISO-8859-1', header = None)
```

```python
clinton_debate_1.columns = ["Clinton_Debate_1"] 
```

```python
clinton_debate_1.head()   #will want to take out 'applause' in first row, but just running this as is to make sure everything works
```

```python
clinton_debate_1.shape   #(92 rows compared to trump's 123 rows)
```

```python
%%time
clinton_debate_1['sentiment_subjectivity'] = clinton_debate_1['Clinton_Debate_1'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
clinton_debate_1.head()     
```

```python
clinton_debate_1['sentiment_polarity'] = clinton_debate_1['Clinton_Debate_1'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
clinton_debate_1.head()    
```

```python
clinton_debate_1['sentiment_subjectivity'].plot()
```

```python
clinton_debate_1['sentiment_polarity'].plot()
```

```python
with open('ClintonPresDebate1.txt', 'r', encoding = "ISO-8859-1" ) as myfile:
    data1=myfile.read()
```

```python
TextBlob(data1).sentiment.subjectivity
#subjectivity (0 to 1): scores closer to 0 are more objective in tone, scores closer to 1 are more subjective in tone
```

```python
TextBlob(data1).sentiment.polarity 
#polarity (-1 to 1): scores closer to -1 are more negative in tone, closer to 0 are more neutral, and closer to 1 are more positive in tone.
```

```python
# In Debate 1:
#     Clinton Subjectivity: 0.4748706812745692 (slightly more objective than Trump overall)
#     Trump Subjectivity: 0.527053192826821
        
#     Clinton Polarity: 0.1679563135448666 (slightly more positive in tone overall)
#     Trump Polarity:  0.12492547933096725
```
